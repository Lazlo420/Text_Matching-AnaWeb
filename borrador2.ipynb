```python
import datetime as dt
import requests
from bs4 import BeautifulSoup
import os
import multiprocessing as mp

start = 736330
registro = {}

def fetch_links(date):
    try:
        url = "https://diariocorreo.pe/archivo/todas/" + date + "/"
        req = requests.get(url)
        soup = BeautifulSoup(req.content, "html.parser")
        links = soup.select("a.story-item__title.block.overflow-hidden.primary-font.line-h-xs.mt-10")
        links = ["https://diariocorreo.pe" + link["href"] for link in links]
        registro[date] = links
        print(date, len(links))
    except Exception as e:
        print(f"Error processing date: {e}")


pool = mp.Pool(processes=10)
results = []
for ordinal in range(start, 737060):
    date = str(dt.date.fromordinal(ordinal))
    result = pool.apply_async(fetch_links, args=(date,))
    print(registro)
    results.append(result)

pool.close()
pool.join()

for result in results:
    result.get()
```

```python
import requests_html
from urllib.parse import urlunparse
import datetime as dt
from concurrent.futures import ThreadPoolExecutor

start = 736330
base_url = "https://diariocorreo.pe/archivo/todas/2017-01-01/"
registro = {}

def fetch_links(date):
    compilao = []
    complete_url = base_url + str(date) + '/'
    session = requests_html.HTMLSession()
    response = session.get(complete_url)
    link_tags = response.html.find('a.story-item__title')
    if not link_tags:
        print(f'revisar {date}')
    else:
        for tag in link_tags:
            link = tag.attrs['href']
            if not link.startswith('https://diariocorreo.pe'):
                link = urlunparse(('https', 'diariocorreo.pe', link, '', '', ''))
            compilao.append(link)
    registro[date] = compilao
    print(f'{date} - {len(compilao)} links')

##--esto si fue claude, no sabia que existia esto xd
with ThreadPoolExecutor() as executor:
    while start < 737060:
        date = dt.date.fromordinal(start)
        executor.submit(fetch_links, date)
        start += 1

with open("links_correo.txt","w") as file:
    for i in registro.keys():
        for j in registro[i]:
            file.write(f'{j}\n')
            
```
