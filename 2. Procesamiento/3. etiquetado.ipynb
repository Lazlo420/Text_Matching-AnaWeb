```python
import pandas as pd

comercio = pd.read_csv("comercio_df1.txt")
correo = pd.read_csv("correo_df1.txt")
p21 = pd.read_csv("peru21_df1.txt")
```

```python
gen=pd.concat([comercio,correo,p21],axis=0)
gen.shape
```

```python
##periodo de tiempo
inicio='2018-08-01'
fin='2018-08-15'
df1=gen[(gen['dia']<=fin)&(gen['dia']>=inicio)]
df1.to_csv("primera_quincena_Agosto_2018.csv")
df1
```
