### Estandarizacion base

- minusculas
- puntuacion
- stopwords
- tokenizacion
- lematizacion

```python
import nltk
import spacy
from nltk.corpus import stopwords
stopwords_es = set(stopwords.words('spanish'))
nlp = spacy.load("es_core_news_sm")
```

```python
def preprocesamiento(texto):
    reemplazo_acentos = {'á': 'a', 'é': 'e', 'í': 'i', 'ó': 'o', 'ú': 'u'}
    
    limpio = texto.strip().lower()
    limpio = re.sub(r'[áéíóú]', lambda x: reemplazo_acentos[x.group()], limpio)
    
    tokens = nltk.word_tokenize(re.sub(r'[^\w\s]', '', limpio), language="spanish")
    
    tokens_sin_stopwords = [word for word in tokens if word not in stopwords_es]
    
    doc = nlp(" ".join(tokens_sin_stopwords))
    lemmas = [token.lemma_ for token in doc]
    
    limpio = " ".join(lemmas)
    return limpio
```
